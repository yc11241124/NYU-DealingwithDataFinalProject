By just looking at the price per night, we can see that the top 5 most expensive neighborhoods are not the same as the top 5 most popular. This could potentially lead us to the conclusion that because there is more competition in the more popular neighborhoods that these hosts need to price their listings more competitively. However, we can also see that the most popular neighborhoods are also not among the 5 least expensive.

###Counting how many times words are used in AirBnB listings in NYC

First, we had to split the listing names from the pandas dataframe into a string.

```python
listing_names = airdata['name'].str.lower().to_string()
listing_names
```

Notice that there are often non-words (numbers, ampersands, and abbreviations) used in listings. It's likely hosts use these to save space if there is a character limit on names.

Using a regex we can easily create a list of word-like substrings. This removes numbers and non-letter characters. We also specified two characters per word to get rid of single letters. We could potentially add more to the character requirement to get rid of the majority of abbreviations, but we didn't want to risk losing too many actual words. 

```python
#using regex to pull only things that look like words from the string
word_pattern = (r'[a-z]{2,}')
words = re.findall(word_pattern,listing_names)
words[:20]
```

Using the nltk library we can get rid of stopwords (such as "and", "the", etc.) and generate a list of unique words. We can use this list to count how many times unique words appear in listings.

```python
#getting a list of only the unique words in listings

nltk.download('stopwords')
setwords = set(words)
setstopwords = set(stopwords.words('english'))
unique_words = setwords-setstopwords
list(unique_words)[:20]
```

Now we have our top 20 words used in NYC listing names! From this list, we decided to analyze the following:
- cozy
- private
- spacious
- sunny
- beautiful

The majority of these words were in our original list as well. We also decided to still analyze luxury to see if there was a price trend associated with this word.

```python
#Top 20 words used in NYC listing names
count_words_list = []

for word in unique_words:
  count_words_list.append(words.count(word))

list_of_words = list(sorted(zip(count_words_list, unique_words),reverse=True))
list_of_words[:20]
```

###Comparing the most popular words to neighborhoods and price.

First, we add columns to the dataframe to count how many listings include the words we are interested in. More words could be added to the substring list if someone was interested in analyzing different words than what we decided to include here.

```python
#adds columns for the words we are interested in exploring to the dataframe and adds a 1 if that word is in the listing name
sub_list=['private', 'luxury', 'cozy', 'sunny', 'beautiful', 'spacious']

for word in sub_list:
  airdata[word] = airdata['name'].str.contains(word, case=False, na=False).astype(int)

airdata.sort_values(by=['cozy'], ascending=False).head()
```
